# Check for null values and drop them

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Take last 10000 rows of the dataset as Tableau dashboard cannot take the entire dataset. Note that the original dataset (after removing null values) has 2049251 rows.

```python
subset_df = df.tail(10000)
```

```python
subset_df.shape
```

# Save the subset to a CSV file

```python
# Save the subset
subset_file_path = "/content/household_power_clean_subset.csv"
subset_df.to_csv(subset_file_path, index=False)

print(f"Subset saved as {subset_file_path}")
```
